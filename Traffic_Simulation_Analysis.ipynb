```python
import matplotlib.pyplot as plt
from nagel_schreckenberg_model import *
```

```python
r = Road(road_size=30, num_cars=(7, 7), max_speed=4, brake_prob=0.1, changing_prob=0.2, num_episodes=160, 
         flux_marker_position=None, traffic_lights_parameters=[{'pos': 15, 'open_time_right': 8, 'open_time_left': 2}], 
         shared_section_position=15, change_side_range=5)
```

```python
def plot_cumulative_arrival_departure(arrival_counts, departure_counts):
    """ Plots cumulative arrivals and departures on the same plot. """
    plt.figure(figsize=(10, 5))
    plt.plot(arrival_counts, label='Cumulative Arrivals', color='blue', marker='o')
    plt.plot(departure_counts, label='Cumulative Departures', color='red', marker='x')
    plt.title('Cumulative Arrival and Departure Over Time')
    plt.xlabel('Episode')
    plt.ylabel('Count')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def plot_cumulative_flux(flux_counts):
    """ Plots cumulative flux over time. """
    plt.figure(figsize=(10, 5))
    plt.plot(flux_counts, label='Cumulative Flux', color='green', marker='^')
    plt.title('Cumulative Flux Over Time')
    plt.xlabel('Episode')
    plt.ylabel('Count')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
flux_data = r.run(terminal_printing=False)  # This should run the simulation and return the flux data
plot_cumulative_flux(flux_data)
```

```python
plot_cumulative_arrival_departure(r.arrival_count, r.departure_count)
```
